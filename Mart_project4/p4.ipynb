```python
'''
this is a regression problem to predict the sales of a shop

'''

import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('train.csv')
data.dropna(how='any', inplace = True)
train, test = train_test_split(data, test_size=0.2, random_state=0)
data.head()
```

```python





print("shape of training data", train.shape)
print("shape of testing data", test.shape)
```

```python
#prediction variable is the Item_Outlet_Sales

train_x = train.drop(columns = ['Item_Outlet_Sales', 'Item_Identifier', 'Item_Fat_Content', 'Item_Type', 'Outlet_Identifier', 'Outlet_Size', 'Outlet_Location_Type', 'Outlet_Type'], axis = 1)

train_y = train['Item_Outlet_Sales']
```

```python
test_x = test.drop(columns = ['Item_Outlet_Sales', 'Item_Identifier', 'Item_Fat_Content', 'Item_Type', 'Outlet_Identifier', 'Outlet_Size', 'Outlet_Location_Type', 'Outlet_Type'], axis = 1)

test_y = test['Item_Outlet_Sales']


```

```python
model = LinearRegression()

model.fit(train_x, train_y)

print('coeff', model.coef_)
print('intercept', model.intercept_)
```

```python
predict_train = model.predict(train_x)
print('\nItem_Outlet_Sales on training data',predict_train)
```

```python
predict_test = model.predict(test_x)
print(predict_test)
```

```python
rmse_train = mean_squared_error(train_y, predict_train)**(0.5)
print(rmse_train)
```

```python
rmse_test = mean_squared_error(test_y, predict_test)**(0.5)
print(rmse_test)
```
